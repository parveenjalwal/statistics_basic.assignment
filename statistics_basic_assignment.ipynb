```python
#7. What is the interquartile range (IQR), and how is it used to detect outliers ?
data3 = [-35, 4,6,8,9,14,39,45,20,26,29,200,350]
percentiles =np.percentile(data3,[0,25,50,75,100])
Q3 = 39
Q1 = 8
# outliers = Q3 - Q1
iqr = Q3 - Q1
# now finding outliers
lf = Q1 - 1.5 * iqr
uf = Q3 + 1.5 * iqr
# so the value byond -38.5 in lowerfance part and values byond 85.5 in upperfance is called outlier . this is how we captured outlier from data3
print("the iqr is :",iqr)
print("lowerfance :",lf)
print("upperfance :",uf)
sns.boxplot(data3)
```
